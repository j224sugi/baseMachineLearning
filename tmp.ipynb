```python
import pandas as pd
import os

fileNames=["LongMethod.csv","FeatureEnvy.csv"]
folder="C:\\Users\\sugii syuji\\machinelearning\\dataset"
saveFolder="C:\\Users\\sugii syuji\\machinelearning\\dataset2"
if not os.path.isdir(saveFolder):
    os.makedirs(saveFolder)
for fileName in fileNames:
    df=pd.read_csv(os.path.join(folder,fileName),encoding_errors="replace")
    oldMetrics=['fanin_y', 'fanout_y', 'wmc_y', 'loc_y', 'returnsQty_y', 'variablesQty_y', 'parametersQty_y', 'loopQty_y', 'anonymousClassesQty_y', 'innerClassesQty_y', 'ATFD_y', 'ATLD_y', 'LAA_y', 'FDP_y', 'line_y']
    df=df[oldMetrics].fillna(0)
    df.to_csv(os.path.join(saveFolder,fileName),index=False)
```
