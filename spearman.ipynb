```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

datasetFolder="C:\\Users\\sugii syuji\\machinelearning\\dataset2"
codeSmell="GodClass.csv"#,"LongMethod.csv"]

df=pd.read_csv(os.path.join(datasetFolder,codeSmell))
df=df.drop(columns=["file","class","type"])
df_corr=df.corr(method="spearman")
df_corr = df_corr.round(2)#1
print(codeSmell)
display(df_corr)

```
